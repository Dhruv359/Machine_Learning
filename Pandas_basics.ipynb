## Sorting/Describing Data

```python
# Summary Statistics 
df.describe()
```

```python
# Sorting Data 
df.sort_values('Name') # Default is Ascending order.
df.sort_values('Defense',ascending=False)
df.sort_values(['Defense','Name','Speed'])
df.sort_values(['Defense','Name','Speed'], ascending=False)
df.sort_values(['Defense','Name','Speed'], ascending=[1,0,1]) # 1 - Ascending , 0 - Descending. 
```

## Making Changes to Data

```python
# Adding a coloumn to the Data 
df['Total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed']

# Drop or Remove a coloumn 
df = df.drop(columns=['Total'])

# Adding Col 
# .iloc[From This Row : To This Row + 1 , From this Col : To this Col +1], (Axis = 1 -> Horizontal)
df['Total'] = df.iloc[ : , 4:10].sum(axis=1)   

df.head()
```

```python
# Rearranging Cols 

#Rearranging total after Type 2 Col. 
cols = list(df.columns)
df = df[cols[0:4] + [cols[-1]] + cols[4:12]]  # -1 is the Last Column(Total), Essentially we are rearranging. 
df.head()

# Switching Attach and Defense. 
df = df[cols[0:6] + [cols[7]] + [cols[6]] + cols[8:]]  # -1 is the Last Column(Total), Essentially we are rearranging. 
df.head()
```

## Saving/Exporting Data


```python
# Saving data to CSV. 
df.to_csv('Modified_Pokemon.csv')

# Save to csv without index col
df.to_csv('Modified_Pokemon2.csv', index = False)

# Save as xlsx file 
df.to_excel('Modified_Pokemon.xlsx', index = False)

# Save to txt 
df.to_csv('Modified_Pokemon.txt', index = False, sep='\t')
```

## Filtering Data 

```python
#Filtering Columns 
df.loc[df['Type 1']=='Grass']
df.loc[(df['Type 1']=='Grass') & (df['Type 2']=='Poison')]
```

```python
# Filtering Data and Resetting Index of generated subset
new_df = df.loc[ (df['Type 1']=='Grass') & (df['Type 2']=='Poison') & (df['HP'] > 70) ]
# Old indexes of filtered values are stored in a new col called index
new_df 

# To reset index and remove the index col.
new_df = new_df.reset_index(drop=True)
new_df
```

```python
#Advanced Filtering 

# filter data to only include Names that contain 'Mega'.  
df.loc[df['Name'].str.contains('Mega')]

# filter data to NOT include Names that contain 'Mega'.  
df.loc[~df['Name'].str.contains('Mega')]

# Using Regex to Filter. 
import re 
# Filter Type 1 col to only include those Fire or Grass. 
df.loc[df['Type 1'].str.contains('Fire|Grass', regex= True)]
df.loc[df['Type 1'].str.contains('fire|grass', regex= True, flags=re.I)]  # Does the same as the above statement but removes Capitalisation senstivity. 
df.loc[df['Name'].str.contains('^pi[a-z]*', regex= True, flags=re.I)]  # Names that start with pi only. 
```

## Conditional Changing 

```python
# Changes all Values of Type 1 == Fire to Flamer
df.loc[df['Type 1'] =="Fire", "Type 1"] = "Flamer"  
df
#Changing it back
df.loc[df['Type 1'] =="Flamer", "Type 1"] = "Fire"
df

#Changes all values of Legendary to True where Type 1 == Fire 
df.loc[df['Type 1'] =="Fire", "Legendary"] = True
df

#Changes all values of Legendary And Generation to test values if Total > 500 
df.loc[df['Total'] > 500, ['Legendary','Generation']] = ['Test 1', 'Test 2']
df
```

## Aggregate Statistics (Groupby)

```python
df = pd.read_csv('Modified_Pokemon2.csv')
```

```python
df.groupby(['Type 1']).mean()
df.groupby(['Type 1']).mean().sort_values('Defense', ascending=False)
df.groupby(['Type 1']).mean().sort_values(['Defense','Attack'], ascending=True)
```

```python
df.groupby(['HP']).sum()
df.groupby(['Type 1','Type 2']).count()
```
